```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def crawl_pta(url):

    result = []
    req = requests.get(url)
    soup = BeautifulSoup(req.text, "lxml")

    news_links = soup.find_all("li",{'data-id':'id-1'})
    # #looping through article link
    for idx,news in enumerate(news_links):
        penulis = ""
        dosen1 = ''
        dosen2 = ''
        news_dict = {}

        # #find url news
        url_news = news.find('a',{'class':'gray button'}).get('href')

        # find news content in url
        req_news =  requests.get(url_news)
        soup_news = BeautifulSoup(req_news.text, "lxml")

        #Judul Jurnal
        judul = soup_news.find('a',{'class':'title'}).text
        # print("title:", title)

        #Penulis, dospem I, dospem II
        dt = soup_news.find_all('div',{'style':'padding:2px 2px 2px 2px;'})
        penulis = dt[0].text.split(':')[1]
        dosen1 = dt[1].text.split(':')[1]
        dosen2 = dt[2].text.split(':')[1]

        #Abstrak
        abstrak = soup_news.find('p',{'align':'justify'}).text

        #wrap in dictionary
        news_dict['Judul'] = judul
        news_dict['Penulis'] = penulis
        news_dict['Dosen Pembimbing I'] = dosen1
        news_dict['Dosen Pembimbing II'] = dosen2
        news_dict['Abstrak'] = abstrak
        result.append(news_dict)

    return result
data = []
for i in range(1,173):
  cr = crawl_pta('https://pta.trunojoyo.ac.id/c_search/byprod/10/%i')
  data = data+cr
print(data)
df = pd.DataFrame(data)

# save dataframe
#df.to_csv('crawling_pta.csv')
```

# Text Cleaning

```python
# Text Cleaning
def cleaning(text):
    # HTML Tag Removal
    text = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});').sub('', str(text))

    # Case folding
    text = text.lower()

    # Trim text
    text = text.strip()

    # Remove punctuations, karakter spesial, and spasi ganda
    text = re.compile('<.*?>').sub('', text)
    text = re.compile('[%s]' % re.escape(string.punctuation)).sub(' ', text)
    text = re.sub('\s+', ' ', text)

    # Number removal
    text = re.sub(r'\[[0-9]*\]', ' ', text)
    text = re.sub(r'[^\w\s]', '', str(text).lower().strip())
    text = re.sub(r'\d', ' ', text)
    text = re.sub(r'\s+', ' ', text)

    # Mengubah text 'nan' dengan whitespace agar nantinya dapat dihapus
    text = re.sub('nan', '', text)

    return text
```

```python
import gensim
from gensim import corpora
from gensim.models import LdaModel
import pandas as pd
import matplotlib.pyplot as plt

# Ubah teks ke dalam format yang cocok untuk Gensim
documents = data['Abstrak_Tokens']

# Membuat kamus (dictionary) dari kata-kata unik dalam dokumen
dictionary = corpora.Dictionary(documents)

# Membuat korpus (bag-of-words) dari dokumen
corpus = [dictionary.doc2bow(doc) for doc in documents]

# Melatih model LDA
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=30)
```

```python
# Menampilkan topik-topik dan kata-kata kunci yang membentuk setiap topik
topics = lda_model.print_topics(num_topics=-1, num_words=len(lda_model.id2word))
for topic in topics:
    print(topic)

#df = pd.DataFrame(topics)
#df.to_csv('Hasil latih Model LDA berdasarkan proporsi dan topiks.csv')
```

# Proporsi Topik Dalam Dokumen

```python
# Membuat DataFrame untuk menampilkan proporsi topik dalam dokumen
document_topic_df = pd.DataFrame()
judul = data['Judul']

for doc in corpus:
    topic_distribution = lda_model.get_document_topics(doc, minimum_probability=0)
    # doc_topic_props = {f"Topik {topic_id + 1}": prop for topic_id, prop in topic_distribution}
    doc_topic_props = {} #mengubah tampilan agar topik di probalility hilang dan ada pada tabel diatasnya
    for topic_id, prob in topic_distribution:
        key = f"Topik {topic_id + 1}"
        doc_topic_props[key] = prob
    document_topic_df = pd.concat([document_topic_df, pd.Series(doc_topic_props)], ignore_index= True, axis=1)

# document_topic_df ['Judul'] = judul
document_topic_df = document_topic_df.transpose()  # Transpose agar topik menjadi kolom

# document_topic_df.insert(document_topic_df.columns.get_loc('Topik 1') , 'Judul', judul)
document_topic_df ['Judul'] = judul
# Menampilkan tabel proporsi topik dalam dokumen
print("Tabel Proporsi Topik dalam Dokumen:")
document_topic_df
#df = pd.DataFrame(document_topic_df)
#df.to_csv('/content/drive/MyDrive/Colab Notebooks/Proporsi Topik dalam Dokumen.csv')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Membaca data teks dari file CSV
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Proporsi Topik dalam Dokumen.csv')

# Menggunakan TF-IDF untuk mengubah teks menjadi representasi numerik
tfidf_vectorizer = TfidfVectorizer(max_df=0.85, max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Judul'])

# Membuat objek K-Means dengan jumlah cluster tertentu (contoh: 3)
kmeans = KMeans(n_clusters=3, n_init='auto')

# Melakukan clustering pada representasi TF-IDF
kmeans.fit(tfidf_matrix)

# Menambahkan label cluster ke dalam data
data['cluster'] = kmeans.labels_

# Menampilkan hasil
print(data)
```
